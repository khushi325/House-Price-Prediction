# Model Building

```python
# Separate input features and target variable

X = df.drop("Price", axis=1)

y = df["Price"]

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
# Split the dataset into training and testing data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data :", X_train.shape)
print("Testing Data :", X_test.shape)
```

```python
# Train Linear Regression model

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

```python
# Predict house prices using Linear Regression

lr_prediction = lr.predict(X_test)
```

```python
# Evaluate Linear Regression model

from sklearn.metrics import r2_score

lr_score = r2_score(y_test, lr_prediction)

print("Linear Regression R² Score :", round(lr_score,4))
```

# Machine Learning Models

```python
# Train Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(
    random_state=42
)

dt.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```

```python
# Predict using Decision Tree

dt_prediction = dt.predict(X_test)
```

```python
# Evaluate Decision Tree model

dt_score = r2_score(y_test, dt_prediction)

print("Decision Tree R² Score :", round(dt_score,4))
```

```python
# Train Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# Predict using Random Forest

rf_prediction = rf.predict(X_test)
```

```python
# Evaluate Random Forest model

rf_score = r2_score(y_test, rf_prediction)

print("Random Forest R² Score :", round(rf_score,4))
```

```python
# Train Gradient Boosting Regressor

from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(
    random_state=42
)

gb.fit(X_train, y_train)

print("Gradient Boosting model trained successfully.")
```

```python
# Predict using Gradient Boosting

gb_prediction = gb.predict(X_test)
```

```python
# Evaluate Gradient Boosting model

gb_score = r2_score(y_test, gb_prediction)

print("Gradient Boosting R² Score :", round(gb_score,4))
```

# Model Evaluation

```python
# Calculate model evaluation metrics

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, rf_prediction)
mse = mean_squared_error(y_test, rf_prediction)
rmse = np.sqrt(mse)

print("Mean Absolute Error :", round(mae,2))
print("Mean Squared Error :", round(mse,2))
print("Root Mean Squared Error :", round(rmse,2))
```
